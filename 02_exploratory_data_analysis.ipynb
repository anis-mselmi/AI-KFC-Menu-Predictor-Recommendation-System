```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

try:
    df = pd.read_csv('cleaned_kfc_data.csv')
    print("Data loaded.")
except FileNotFoundError:
    print("Error: 'cleaned_kfc_data.csv' not found.")
```

```python
sns.set_theme(style="whitegrid", palette="pastel")
if 'df' in locals():
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(y='Primary_Item', data=df, order=df['Primary_Item'].value_counts().index, hue='Primary_Item', legend=False)
    plt.title('Popularity of Primary Menu Items', fontsize=16)
    plt.xlabel('Count')
    plt.ylabel('Menu Item')
    plt.show()
```

```python
if 'df' in locals():
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Time_Of_Day', data=df, order=['Lunch', 'Afternoon Snack', 'Dinner', 'Late Night'], hue='Time_Of_Day', legend=False)
    plt.title('Distribution of Orders by Time of Day', fontsize=16)
    plt.xlabel('Time of Day')
    plt.ylabel('Number of Orders')
    plt.show()
```

```python
if 'df' in locals():
    plt.figure(figsize=(10, 5))
    sns.histplot(df['Customer_Age'], kde=True, bins=20, color='orange')
    plt.title('Customer Age Distribution', fontsize=16)
    plt.xlabel('Age')
    plt.show()
```
