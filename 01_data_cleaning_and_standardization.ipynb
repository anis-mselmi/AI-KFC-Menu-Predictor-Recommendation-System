```python
import pandas as pd
import numpy as np
import random
random.seed(42)
np.random.seed(42)
print("Libraries imported.")
```

```python

num_orders = 500
primary_items = ['Zinger Burger', 'Original Recipe Chicken', 'Popcorn Chicken', 'Twister Wrap', 'Hot Wings', 'zinger burger', 'ZINGER BURGER', 'Original Recipe', None]
side_items = ['Fries', 'Coleslaw', 'Mashed Potatoes', 'Corn on the Cob', 'Biscuits', 'FRIES', None, 'mashed potatoes']
times_of_day = ['Lunch', 'Dinner', 'Late Night', 'Afternoon Snack', 'lunch', 'DINNER']

data = {
    'OrderID': list(range(1, num_orders + 1)),
    'Primary_Item': [random.choice(primary_items) for _ in range(num_orders)],
    'Side_Item': [random.choice(side_items) for _ in range(num_orders)],
    'Time_Of_Day': [random.choice(times_of_day) for _ in range(num_orders)],
    'Customer_Age': [random.randint(18, 65) if random.random() > 0.1 else None for _ in range(num_orders)]
}

df = pd.DataFrame(data)
print("Raw Data Sample:")
display(df.head(10))
print(f"\nShape: {df.shape}")
print(f"\nMissing Values:\n{df.isnull().sum()}")
```

```python

text_cols = ['Primary_Item', 'Side_Item', 'Time_Of_Day']
for col in text_cols:
    df[col] = df[col].astype(str).str.lower().str.title()
    df[col] = df[col].replace({'None': np.nan, 'Nan': np.nan})

item_mapping = {
    'Original Recipe': 'Original Recipe Chicken',
    'Zinger Burger': 'Zinger Burger' 
}
df['Primary_Item'] = df['Primary_Item'].replace(item_mapping)

for col in ['Primary_Item', 'Side_Item', 'Time_Of_Day']:
    if not df[col].mode().empty:
        df[col] = df[col].fillna(df[col].mode()[0])

median_age = df['Customer_Age'].median()
df['Customer_Age'] = df['Customer_Age'].fillna(median_age)

print("Data Standardized and Imputed.")
print(f"\nMissing Values after cleaning:\n{df.isnull().sum()}")
display(df.head())
```

```python
df.to_csv('cleaned_kfc_data.csv', index=False)
print("File 'cleaned_kfc_data.csv' saved successfully.")
```
